```python
#extract file
import zipfile

# Check if the file is a valid ZIP file
if zipfile.is_zipfile('orders.csv.zip'):
    # Open the ZIP file
    with zipfile.ZipFile('orders.csv.zip', 'r') as zip_ref:
        # Extract all contents into the current directory
        zip_ref.extractall()
else:
    print("Not a valid ZIP file!")
```

```python
# read the csv file
#read data
#handle null values
import pandas as pd
df= pd.read_csv('orders.csv')
df.head(20)
```

```python
#here it shows the unique values that ship mode have
df['Ship Mode'].unique()
```

```python
#values such as not available, unknown are treated as NAN
df= pd.read_csv('orders.csv', na_values = ['Not Available', 'unknown'])
df.head(20)
```

```python
# rename columns: make lower case and replace space( ) with underscore(_)
df.columns
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.head()
```

```python
# derive new columns discount, sale price and profit
df['discount'] = df['list_price']*df['discount_percent']/100
df.head()
```

```python
df['sale_price'] = df['list_price']-df['discount']
df.head()
```

```python
df['profit'] = df['sale_price']-df['cost_price']
df.head()
```

```python
df.dtypes
# here order_date is written in YYMMDD but dtype is object so we'll convert that
```

```python
#convert data type of order_date
df['order_date'] = pd.to_datetime(df['order_date'],format = "%Y-%m-%d")
df.head()
```

```python
df.dtypes #here dtype finally converted
```

```python
# drop cost_price, list_price and discount_percent
df.drop(columns = ['list_price', 'cost_price', 'discount_percent'], inplace = True)
df.head()
```
